```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
M = 5000
K = 100
length = K * M

data = np.fromfile('output.bin', dtype=float, sep = ',')
print(data.shape)
data = data[:length]

shape = (K, M)
data = np.reshape(data, shape)
print('data shape is', data.shape)
```

```python
plt.plot(data[0, ])
plt.grid()
```

```python
k = np.arange(0, K)
m = np.arange(0, M, 50)
def arr_U(k, m):
    return data[k, m]
fig = plt.figure(figsize = [8, 8])
ax = plt.axes(projection='3d')
X = np.linspace(0, 1 - 1 / M, K)
T = np.linspace(0, 1 - 1 / K, K)
X, T = np.meshgrid(X, T)
k, m = np.meshgrid(k, m)
data_plot = arr_U(k, m)

ax.plot_surface(T, X, data_plot, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
```
