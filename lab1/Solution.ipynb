```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
M = 5000
K = 100
length = K * M

data = np.fromfile('output.bin', dtype=float, sep = ',')
print(data.shape)
data = data[:length]

shape = (K, M)
data = np.reshape(data, shape)
print('data shape is', data.shape)
```

```python
plt.plot(np.linspace(0, 1 - 1 / M, M), data[0, ])
plt.grid()
plt.title("Начальное условие U0(x) = U(x, t=0)")
```

```python
k = np.arange(0, K)
m = np.arange(0, M, 50)
def arr_U(k, m):
    return data[k, m]
fig = plt.figure(figsize = [8, 8])
ax = plt.axes(projection='3d')
X_0 = np.linspace(0, 1 - 1 / M, K)
T_0 = np.linspace(0, 1 - 1 / K, K)
X, T = np.meshgrid(X_0, T_0)
k, m = np.meshgrid(k, m)
data_plot = arr_U(k, m)

ax.plot_surface(T, X, data_plot, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax.set_title('Solution of the equation')
ax.set_xlabel('t')
ax.set_ylabel('X')
ax.set_zlabel('U')
```
