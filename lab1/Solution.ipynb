```python
import numpy as np
import matplotlib.pyplot as plt
M = 1000
K = 1000
D_k = 10
K_d = int(K / D_k)
length = M * K_d

data = np.fromfile('output.bin', dtype=float, sep = ',')
data = data[:length]

shape = (K_d, M)
data = np.reshape(data, shape)
print('data shape is', data.shape)
```
