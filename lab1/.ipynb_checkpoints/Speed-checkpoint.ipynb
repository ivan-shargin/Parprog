```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
speed_arr = np.array([[1.177617, 0.644233, 0.430911, 0.326189, 0.530182],
                      [1.187340, 0.632068, 0.417151, 0.313215, 0.539709],
                      [1.178490, 0.613001, 0.430980, 0.306455, 0.528856]])
size = np.arange(1, 6)
```

```python
speed_arr = np.mean(speed_arr, axis=0)
print(speed_arr)
```

```python
plt.figure(figsize=[5, 3])
plt.title('Speed of comutation')
plt.xlabel('Number of processes')
plt.ylabel('Time, s')
plt.tight_layout()
plt.grid()
plt.plot(size, speed_arr)
```
